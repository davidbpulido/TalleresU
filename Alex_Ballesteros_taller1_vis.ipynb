### Visualización 1 

### What?,Why?,How?

  * **what** 
    + **Data types** El tipo de los datos considero que son items que se le ponen a una posición en el cuerpo.
    + **Dataset types** El tipo de base de datos creo que puede ser geométrico, ya que representa valores en zonas del cuerpo y el area que ellas comprenden.
    + **Data and dataset types** Son fields ya que contienen el item, el nombre de la ubicación y la posición de estos.
    + **Dataset Availability** Es estático ya que presenta unos valores para un momento de tiempo particular(como si le tomaramos una foto en el momento y ya).

    +  **Attribute types** Los datos son de orden cuantitativo y no presentan ordenación.


* **Why?**
    + **Actions** Present->Lookup->Compare

    + **Targets** Shape
    

* **How?**
    + **Encode** Use
    


### Marcas y canales

 + Marca de áreas
 + Posición horizontal y vertical para el item
 + Tamaño del área
    

### Reglas Generales

Me parece que cumple con las reglas generales sobre todo por lo resumidos que están los datos teniendo en cuenta que son 6 deportes distintos y lo fácil que es para el usuario interpretar los datos, se hubiera podido hacer un 3d, pero es un exceso de medios así que el 2d me parece bien utilizado.

### Visualización 2

### What?,Why?,How?

  * **what** 
    + **Data types** El tipo de los datos son atributos.
    + **Data and dataset types** Es una tabla que contiene un item de cantidad y un atributo para el mismo.
    + **Dataset types** El tipo de base de datos es una tabla.
    + **Dataset Availability** Es estático ya que presenta unos valores para un momento de tiempo particular(como si le tomaramos una foto en el momento y ya).

    +  **Attribute types** Los datos son de orden cuantitativo y presentan ordenación secuencial.


* **Why?**
    + **Actions** Present->Locate->Summarize

    + **Targets** Features
    

* **How?**
    + **Encode** Order
    

### Marcas y canales

 + Marca de áreas
 + Posición horizontal y vertical para el atributo cuantitativo
 + Tamaño del área
 + Usa un canal de color en un tono de amarillo, naranja, rojo

### Reglas Generales

Por la manera en que se presentan lo datos me parece que cumple con las reglas generales.
