```python
import pandas as pd
import numpy as np
import seaborn as sns
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
import os
```

```python
path='C:/Users/David/Downloads'
fullp=os.path.join(path, 'database.xlsx')
datos=pd.read_excel(fullp)
datos=datos.dropna() ## eliminamos los datos nulos
```

Como se planea utilizar las variables "continent" y "domain", que nos indican el continente del cual es procedente el famoso correspodiente y el área de mayor contribución, con el fin de intentar reconocer alguna relación entre estas, limpiaremos la variable "continent" que contiene unos pocos NAs y no se hará más limpieza a la base.

### Visualización

```python
df = pd.DataFrame({ 'from':datos.continent, 'to':datos.domain})
plt.figure(figsize=(15,15))
G=nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph() )
nx.draw(G, with_labels=True)
plt.title("Grafo dirigido continente y domain")
plt.show()
```

### What?,Why?,How?

    + **Data and dataset types** Son redes, contiene los items y la conexión.
