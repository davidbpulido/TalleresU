### What?,Why?,How?

  * **what** 
    + **Data types** El tipo de los datos son items que se relacionan entre ellos.
    + **Dataset types** El tipo de base de datos es una red, que se presenta en dos columnas y el enlace será definido de columna a columna cuando las filas sean igual.
    + **Data and dataset types** Son arboles y redes, contiene los items y la conexión.
    + **Dataset Availability** Es estático ya que presenta unos valores para un momento de tiempo particular(como si le tomaramos una foto en el momento y ya).

    +  **Attribute types** Los datos son nodos y se presenta una poca cantidad de ellos

La visualización nos desea mostrar un gráfico de árbol para mostrarnos la conexión entre los profesores y las materias que dictan en la facultad de estadística, estos datos se obtienen empiricamente, y se presentan todas las materias que el profesor haya dictado o esté dictando desde el periodo 2017-1(periodo de ingreso mio a la facultad), hay que tener en cuenta que solo están las materias que son de la facultad y no se tuvieron en cuenta electivas, por la estructura de los datos las conexiones se van a dar siempre de profesor<->materia y nunca serán materia<->materia o profesor<->profesor.
* **Why?**
    + **Actions** Present->Locate/Explore->Identify
    + **Network Data** Topology/Paths
    
Se hace a modo de "descubrimiento" y de alguna manera se puede mirar preferencias hacia ciertas áreas por parte de algunos profesores, o ver la carga acádemica de algunos profesores en cuanto a materias dictadas, se puede ver como las materias Cálculo Actuarial y Seguros es unicamente dictada por la profesora Deisy, y en cuanto a algunos profesores como Wilmer y Dagoberto vemos mucha acumulación de conexiones.
* **How?**
    + **Encode** Express
  
Se creo un archivo de excel con dos columnas "From" y "To" y se generó el gráfico con Python3.

### Aporte con respecto al grafico

El gráfico se puede mejorar en cuanto a su forma de muchas formas, se podría generar una variable cuantitativa y de esta manera podriamos agregarle un canal de área y de esta manera identificar acumulaciones de conexiones, tambien se podría hacer una distinción para las materias según su área más común de desempeño (análitica, muestreo, bayesiana, finanzas), tambien se podría considerar reformar la base de datos, obtener los datos exactos de las materias y los profesores que la dictaban por periodo y generar un gráfico interactivo en el cual se pudiera ver como ha cambiado a lo largo de los periodos.
